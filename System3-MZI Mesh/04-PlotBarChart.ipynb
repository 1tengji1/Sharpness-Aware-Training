```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

accu_off = 68.6
accu_nat = 79.8
accu_pat = 70.2
accu_dat = 94.6
accu_sam = 92.3
accu_pat_sam = 97.1


time_off = 356.1
time_nat = 338.2
time_pat = 393.8
time_dat = 1272.7
time_sam = 513.8
time_pat_sam = 623.1


lambda_off = 93.5
lambda_nat = 96.8
lambda_pat = 114.7
lambda_dat = 252.4
lambda_sam = 0.3
lambda_pat_sam = 0.6


# labels = ["Offline", "NAT", "PAT", "DAT", "SAM", "PAT+SAM"]
labels = ["label", "label", "label", "label", "label", "label"]

wo_color = 'salmon'       # Accuracy 
w_color = 'steelblue'     # Time 
l_color = 'limegreen'     # λ 

fig, ax = plt.subplots(figsize=(6, 3), dpi=1000)

x = np.arange(6) + 0.5

# y1, y2, y3 --- Accuracy、Time、lambda
y1 = [100-accu_off, 100-accu_nat, 100-accu_pat, 100-accu_dat, 100-accu_sam, 100-accu_pat_sam]
y2 = [time_off, time_nat, time_pat, time_dat, time_sam, time_pat_sam]
y3 = [lambda_off, lambda_nat, lambda_pat, lambda_dat, lambda_sam, lambda_pat_sam]

bar_width = 0.15

ax.bar(
    x - bar_width/2 - 0.02,
    y1,
    bar_width,
    facecolor=wo_color,
    edgecolor='white',
    alpha=0.5,
    label='Accuracy'
)

ax.set_xlim(0, 6.3)    
ax.set_xticks(x, labels)

ax.set_yticks([0, 15, 30], ['0', '0', '0'])
ax.tick_params(axis='y', labelcolor=wo_color)
ax.set_ylabel('Error rate (%)',color=wo_color)
ax1 = ax.twinx()
ax1.bar(
    x + bar_width/2 + 0.02,
    y2,
    bar_width,
    facecolor=w_color,
    edgecolor='white',
    alpha=0.5,
    label='Time'
)

ax1.set_ylim(0, 1600)
ax1.set_yticks([0, 750, 1500], ['0', '0', '0'])
ax1.tick_params(axis='y', labelcolor=w_color)
ax1.set_ylabel('Training time (s)', color=w_color)
ax2 = ax.twinx()
ax2.spines["right"].set_position(("outward", 60))  
ax2.bar(
    x + 3*bar_width/2 + 0.04,
    y3,
    bar_width,
    facecolor=l_color,
    edgecolor='white',
    alpha=0.5,
    label=r'$\lambda$'
)

ax2.tick_params(axis='y', labelcolor=l_color)
ax2.set_yscale('log')
ax2.set_yticks([1, 10, 100],['0','0','0'])
ax2.set_ylabel('Sentivity',color=l_color)
lines1, labels1 = ax.get_legend_handles_labels()
lines2, labels2 = ax1.get_legend_handles_labels()
lines3, labels3 = ax2.get_legend_handles_labels()

plt.tight_layout()
plt.show()
```
